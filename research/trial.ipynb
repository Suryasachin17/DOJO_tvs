```python
# import cv2
# import numpy as np
# import flask as Flask






# # Create a blank canvas for the signature pad
# canvas_width, canvas_height = 800, 400
# canvas = np.ones((canvas_height, canvas_width, 3), dtype=np.uint8) * 255  # White background
# drawing = False  # True if the user is drawing

# # Variables to store previous coordinates
# prev_x, prev_y = -1, -1

# def draw_signature(event, x, y, flags, param):
#     global drawing, prev_x, prev_y

#     # Mouse event handlers
#     if event == cv2.EVENT_LBUTTONDOWN:  # User starts drawing
#         drawing = True
#         prev_x, prev_y = x, y
#     elif event == cv2.EVENT_MOUSEMOVE and drawing:  # User is drawing
#         if prev_x != -1 and prev_y != -1:
#             cv2.line(canvas, (prev_x, prev_y), (x, y), (0, 0, 0), thickness=2)  # Draw black line
#         prev_x, prev_y = x, y
#     elif event == cv2.EVENT_LBUTTONUP:  # User stops drawing
#         drawing = False
#         prev_x, prev_y = -1, -1

# # Setup OpenCV window and mouse callback
# cv2.namedWindow("Signature Pad")
# cv2.setMouseCallback("Signature Pad", draw_signature)

# app = Flask(__name__)
# @app.route("/signature",method = ["Get","post"])
# def signature():
#     while True:
#         cv2.imshow("Signature Pad", canvas)
#         key = cv2.waitKey(1)

#         if key == ord('s'):  # Press 's' to save the signature
#             filename = "signature.png"
#             cv2.imwrite(filename, canvas)
#             print(f"Signature saved as {filename}")
#             break
#         elif key == ord('c'):  # Press 'c' to clear the canvas
#             canvas[:] = 255
#             print("Canvas cleared")
#         elif key == 27:  # Press 'ESC' to exit without saving
#             print("Exiting without saving")
#             break

#     cv2.destroyAllWindows()



# if __name__ == "__main__":
    
#     app.run(host="0.0.0.0", port=5000,  use_reloader=False)

import cv2
import numpy as np
from flask import Flask, request, jsonify, render_template
import base64

app = Flask(__name__)

# Global canvas for signature
canvas = np.ones((400, 800, 3), dtype=np.uint8) * 255  # White background
drawing = False
prev_x, prev_y = -1, -1

def draw_signature(event, x, y, flags, param):
    global drawing, prev_x, prev_y, canvas

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        prev_x, prev_y = x, y
    elif event == cv2.EVENT_MOUSEMOVE and drawing:
        if prev_x != -1 and prev_y != -1:
            cv2.line(canvas, (prev_x, prev_y), (x, y), (0, 0, 0), thickness=2)
        prev_x, prev_y = x, y
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        prev_x, prev_y = -1, -1

@app.route('/')
def index():
    return render_template('signature.html')

@app.route('/signature', methods=['GET', 'POST'])
def signature():
    if request.method == 'POST':
        # Save signature from base64 data
        signature_data = request.json.get('signature')
        if signature_data:
            # Remove the "data:image/png;base64," part if it exists
            if ',' in signature_data:
                signature_data = signature_data.split(',')[1]
            
            # Decode base64 and save
            with open("signature.png", "wb") as f:
                f.write(base64.b64decode(signature_data))
            return jsonify({"status": "Signature saved successfully"})
        return jsonify({"error": "No signature data"}), 400

    # GET method for retrieving signature
    return jsonify({"message": "Send POST request with signature"})

def start_signature_pad():
    global canvas
    cv2.namedWindow("Signature Pad")
    cv2.setMouseCallback("Signature Pad", draw_signature)

    while True:
        cv2.imshow("Signature Pad", canvas)
        key = cv2.waitKey(1)

        if key == ord('s'):  # Save
            filename = "signature.png"
            cv2.imwrite(filename, canvas)
            print(f"Signature saved as {filename}")
            break
        elif key == ord('c'):  # Clear
            canvas[:] = 255
            print("Canvas cleared")
        elif key == 27:  # Exit
            print("Exiting without saving")
            break

    cv2.destroyAllWindows()

if __name__ == "__main__":
    # Option to start signature pad manually
    start_signature_pad()
    
    # Uncomment the following to run Flask server
    app.run(host="0.0.0.0", port=5000, debug=True)

```
